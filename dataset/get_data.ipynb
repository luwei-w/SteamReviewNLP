```python
import steamreviews
import pandas as pd 
```

```python
columns = ['game_id','game_name','recommendation_id','user_id','num_games_owned',
           'num_reviews','playtime_forever','playtime_at_review','review','timestamp_updated'
           ,'voted_up','votes_up','votes_funny','weighted_vote_score','comment_count',
           'steam_purchase','received_for_free','written_during_early_access']
```

```python
id_list = {1097150:'Fall Guys',1085660:'Destiny 2',271590:'Grand Theft Auto V'
           ,289070:'Sid Meier\'s Civilization VI',646570:'Slay the Spire'
           ,105600:'Terraria',252490:'Rust',255710:'Cities: Skylines',1100600:'Football Manager 2020'
           ,230410:'Warframe',220:'Half-Life 2',383120:'Empyrion - Galactic Survival'
           ,1089350:'NBA 2K20',292030:'The Witcher 3: Wild Hunt',1174180:'Red Dead Redemption 2',
           322330:'Don\'t Starve Together',548430:'Deep Rock Galactic',812140:'Assassin\'s Creed Odyssey',
           107410:'Arma 3',374320:'DARK SOULS™ III',582010:'Monster Hunter: World',381210:'Dead by Daylight',
          4000:'Garry\'s Mod',1190460:'DEATH STRANDING',359550:'Tom Clancy\'s Rainbow Six Siege'}
#25 games, 1000 reviews per game
#https://steamdb.info/
```

```python
#only english reviews required
#https://partner.steamgames.com/doc/store/getreviews

#voted_up -> pos/neg 
# if no update, then timestamp_updated = timestamp_created
#all true/false convert to 0/1
r = []
for app_id,app_name in id_list.items():
#     request_params = dict()
#     # Reference: https://partner.steamgames.com/doc/store/getreviews
#     request_params['filter'] = 'all'  # reviews are sorted by helpfulness instead of chronology
#     request_params['day_range'] = '50'
    review_dict, query_count = steamreviews.download_reviews_for_app_id(app_id)
    count = 0
    for k,v in review_dict['reviews'].items():
        if v['language']=='english':
            q=[app_id,app_name,v['recommendationid'],v['author']['steamid'],v['author']['num_games_owned']
                      ,v['author']['num_reviews'],v['author']['playtime_forever'],v['author']['playtime_at_review']
                     ,v['review'],v['timestamp_updated'],v['voted_up'],v['votes_up'],v['votes_funny']
                      ,v['weighted_vote_score'],v['comment_count'],v['steam_purchase'],v['received_for_free']
                     ,v['written_during_early_access']]
            for i in range(len(q)):
                if q[i] == True:
                    q[i] = 1
                elif q[i] == False:
                    q[i] = 0
            r.append(q)
            count += 1
#         if count == 2000:
#                 break
```

```python
df = pd.DataFrame(r, columns = columns) 
```
